```python
import numpy as np 
import pandas as pd 
import os 
import time 

from selenium import webdriver
from selenium.webdriver.common.by import By   # 특정 요소 존재 여부 확인
from selenium.webdriver.support.ui import WebDriverWait # 웹 페이지 로딩 완료될 때까지 대기
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys # key event 발생 시킴

# Exception
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
ce = pd.read_csv('./data/Daum_BAZAAR_Ce_contents_processed.csv', sep = "|" )
ce2 = pd.read_csv('./data/Daum_BAZAAR_Ce2_contents_processed.csv', sep = "|" )
```

# ce 컬럼 순서 바꾸기 


```python
ce = ce[['title','date','link','category','title_brand','check','why','Brand','content']]
```

```python
ce.head()
```

```python
ce2
```

# 셀린, 셀린느  병합 

```python
ce_union = pd.concat([ce, ce2])
```

```python
ce_union
```

## 중복값 확인 

```python
isDuple = ce_union.duplicated(['title'], keep = 'last')
```

```python
isDuple_df = pd.DataFrame(isDuple)
isDuple_df
```

```python
isDuple_df[0].value_counts()
```

## 중복값 제거 

```python
ce_union_noDuple = ce_union.drop_duplicates(['title'], keep='first')
```

```python
ce_union_noDuple
```

### Brand column = Ce 로 통일 

```python
ce_union_noDuple['Brand'] = "Ce"
ce_union_noDuple['Brand'].value_counts()
```

# CSV 저장 

```python
ce_union_noDuple.to_csv('./data/Daum_BAZAAR_Ce_Union_contents_processed.csv', sep = "|", encoding = 'utf-8')
```
