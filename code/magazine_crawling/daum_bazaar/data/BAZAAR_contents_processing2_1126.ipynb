### csv 로드

```python
b_list = [df_Ch, df_Ce, df_D, df_YSL]

Ch_list  =   ['샤넬','샤낼', 'Chanel']
Ce_list  =   ['셀린느','셀린','샐린','샐린느','Celine']
D_list   =   ['디올','디오르', '크리스찬 디올','크리스챤 디올','크리스찬디올','크리스챤디올', 'Dior']
YSL_list =   ['로랑', '생로랑','생 로랑', '입생로랑','이브 생 로랑','YSL', 'Saint Laurent', 'Yves Saint Laurent']

total_list = [Ch_list, Ce_list, D_list, YSL_list]
```

# test

## content 본문내용 맨앞에 카테고리 글자 삭제하기 & 리스트화 

```python
for idx, dataframe in enumerate(b_list):  # 각 브랜드 데이터, idx : 0~ 3 
    k = dataframe['content']           # 각 브랜드의 content 시리즈 가져옴 
    for i in range(len(dataframe)):  # 데이터 프레임 각 행 
        temp_list = list(k[i].split("\',"))
        temp_list2 = temp_list[1:]
        dataframe['content'][i] = temp_list2
```

### title_brand 적용 

```python
for idx, dataframe in enumerate(b_list):  # 각 브랜드 데이터, idx : 0~ 3 
    for i in range(len(dataframe)):  # 데이터 프레임 각 행 
        for word in total_list[idx]: # 
            if word in dataframe['title'][i]:
                dataframe['title_brand'][i] = 1
                print("적용!!!!  :  ", dataframe.iloc[i])
```
